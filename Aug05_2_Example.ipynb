```python
# 필요한 라이브러리 불러오기
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# http://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=cs212
# 전체 내용 데이터를 크롤링하기
# parser 종류 : html.parser => html5lib

res = requests.get('http://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=cs212')
menu = BeautifulSoup(res.content, 'html5lib')
print(menu.prettify())
```

```python
# 순위, 상호, 업종, 주소, url에 대한 내용을 출력

# select() vs select_one()
# select() : 매칭되는 여러 값들을 리스트에 담음
# select_one() : 매칭되는 값 하나만 가져와서 => 객체
# select관련 => 하나의 열에 여러 선택자가 필요한 경우
#   => .000 or .XXX => (or: ','(쉼표))

# 속성값 구해야 할때 : [ ]를 사용 !!

ul = menu.select_one('body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul')

for li in ul.select('li'):
  print('순위 :', li.select_one('p').string)
  # print('순위: ', li.select_one('.numTop, .rankNum').string)
  print('상호 :', li.select_one('.restName').string)
  print('업종 :', li.select_one('.listType').string)
  print('주소 :', li.select_one('.listArea').string)
  print("-----------------------------------")
  #print('url :', li.select_one('.listPhoto a')['href'])
  print('url :', li.select_one('.restName a')['href'])
  print("-----------------------------------")
  print('==' * 30)



```

```python
# 위의 데이터들을 list로 나타내는 작업

#[ {'순위' : '1', '상호' : 'xxxx', '업종' : '한식', ...},
#{'순위' : '1', '상호' : 'xxxx', '업종' : '한식', ...},
#{'순위' : '1', '상호' : 'xxxx', '업종' : '한식', ...}]
menuList = []
urlList =[]
for li in ul.select('li'):
  menudict={}
  menudict['순위'] = li.select_one('p').string
  menudict['상호'] = li.select_one('.restName').string
  menudict['업종'] = li.select_one('.listType').string
  menudict['주소'] = li.select_one('.listArea').string
  menuList.append(menudict)
for m in menuList:
    print(m)
# url에 대한 내용은 따로 별개의 list에 넣기

for li in ul.select('li'):
  urldict={}
  urldict['u'] = li.select_one('.restName a')['href']
  urlList.append(urldict)
for u in urlList:
    print(u)
print(urlList)
```

```python
# menuList를 가지고 순위를 index로 하는 DataFrame을 생성
df = pd.DataFrame(menuList)
df.set_index('순위', inplace=True)
df
```

```python
#http://www.menupan.com/ 이 기본 주소값 뒤에 urlList에 담아놓은 주소값 붙여서 요청
# 전화번호, 상세주소 (두개 중 하나)를 가져온 후에
# 만들어놓은 DataFrame에 붙이기(+ '주소' 열은 삭제)
baseUrl = 'http://www.menupan.com'
telList = []
addrList = []

for u in urlList:
  res = requests.get(baseUrl+u)
  soup = BeautifulSoup(res.content, 'html5lib')
  # print(soup.select_one('.tel1').string)
  # print(soup.select_one('.add1 a').string)
  telList.append(soup.select_one('.tel1').string)
  addrList.append(soup.select_one('.add1 a').string)

df['전화번호'] = telList
df['상세주소'] = addrList
df.drop('주소', axis=1, inplace=True)
df
```

```python
# 업종이 한식인 가게의 정보만 출력
df[df['업종']] == '한식'
```

```python
# 특정 구에 있는 가게의 정보만 출력
# Pandas에서 Series의 특정 문자열 찾기 : str.contains()

df[df['상세주소'].str.contains('양천구')]
```
