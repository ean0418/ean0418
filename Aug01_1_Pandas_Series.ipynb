# 1. 시리즈 (Series)

Series : 객체를 담을 수 있는 1차원 배열스러운 자료구조



```python
# 1-1. 딕셔너리로 만들기 (dict)
# 시리즈는 딕셔너리 구조와 비슷함
dic = {'a':1, 'b':2, 'c':3, 'd': 4}
series_dic = pd.Series(dic)

print(type(series_dic))
series_dic
# 딕셔너리의 key가 인덱스(index)로, value가 값으로
```

```python
# 인덱스와 값은 아래처럼 확인
print(series_dic.index)
print(series_dic.values)
```

```python
# 직접 만들기도 O
# 1-2. 리스트로 만들기

obj = pd.Series([4, 5, -4, -5])
obj

print(obj.index)
print(obj.values)

# RangeIndex(start=0, stop=4, step=1)
# 시리즈를 만들 때, 따로 인덱스를 입력하지 않았기 때문에
#   0부터 시작하는 인덱스가 부여(start=0)
#   4개의 인덱스 (stop = 4) / 1씩 건너서 (step = 1)
```

```python
obj = pd.Series([4, 5, -4, -5], index = ['a', 'b', 'c', 'd'])
obj
```

```python
# 1-3. 튜플로 만들기
tup_data = ('Ean', '2024-04-18','남', True)
series_data = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])
series_data
```

```python
# 1-4. 요소 선택 / 인덱싱
print(series_data[0])
print(series_data['이름'])
```

```python
# 여러 값을 인덱싱할 때에는 위에처럼 값만 반환하지 않고,
# 짝을 이루는 인덱스와 값을 모두 반환

# **************주의할 점
# 각각의 위치 or 이름으로 인덱싱할때는 대괄호 2개를 씌워줘야 함... [[ ]]

print(series_data[[0, 2]])
print(series_data[['이름', '성별']])
```

```python
# : 로 인덱싱 할 떄는 대괄호 하나만 필요
print(series_data[1:3])
print(series_data['생년월일':'성별'])
# 인덱스의 이름으로 사용하면 범위 끝('성별')이 포함됨
```

```python
# Boolean배열을 사용해서 값 걸러내기 or 산술 연산 or 수학 함수 적용 O
obj2 = pd.Series([4, 5, -4, -5], index = ['a', 'b', 'c', 'd'])
print(obj2[obj2 > 0])
print()
print(obj2 * 2)
print()
print('b' in obj2)
print()
print('e' in obj2)
```

# 2. 데이터프레임 (DataFrame)

데이터프레임은 시리즈가 여러개 합쳐진 자료형

표, 엑셀과 같은 스프레드시트 형식의 자료구조

다양한 컬럼이 있고, 각각의 컬럼안에는 서로 다른 종류의 자료형(숫자, 문자열, 논리,...) 이 담김

DataFrame : row(행), column(열)에 대한 index를 가지고 있음

```python
# 2-1. 딕셔너리로 만들기
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2' : [7,8,9], 'c3' : [10,11,12]}
df = pd.DataFrame(dict_data)

print(type(df))
df
# 1개의 열들이 1개의 시리즈, key값이 열 이름
```

```python
# 2-2. 리스트로 만들기 1
#   같은 길이의 리스트에 담긴 dict or numpy를 이용

data = {'time' : ['moring','afternoon', 'night'],
        'startHour' : [6, 12, 18],
        'behavior' : ['getup', 'lunch', 'gohome']}
data1 = pd.DataFrame(data)
data1
```

```python
# 원하는 순서대로 columns을 지정하면, 그 순서대로 df객체가 생성
pd.DataFrame(data, columns = ['time', 'behavior', 'startHour'])
```

```python
# 2-2. 리스트로 만들기 2
df = pd.DataFrame([[21,'남','00대'],[22, "여",'ㅁㅁ대']],
                  index=['황이안','김이안'],
                  columns=['나이','성별','학교'])
print(df)
print(df.index)
print(df.columns)
df
```

```python
# 2-3. 행, 열의 이름 변경 : 기본적인 방식
df.index = ['student1','studnet2']
df.columns = ['나이','성별','학교']
df

# 인덱스(index)와 열(column) 이름의 수에 맞게 지정해주기!
```

```python
# 2-4. 행, 열 이름 변경 : rename() 함수
# df.rename(column / index = {변경 전 이름 : 변경 후 이름,...}, inplace=True)
# 로 사용할 수 있고,
# inplace = True 옵션을 지정해주면 df에 바로 적용 됨
df = pd.DataFrame([[21,'남','00대'],[22, "여",'ㅁㅁ대']],
                  index=['황이안','김이안'],
                  columns=['나이','성별','학교'])
df
```

```python
# 열 이름 변경
df.rename(columns = {'나이' : 'age', '성별' : 'gender', '학교' : 'school'}, inplace=True)
df
```

```python
# 행 이름 변경
df.rename(index = {'황이안' : 'student1', '김이안' : 'student2'}, inplace=True)
df
```

```python
# 행 삭제 : drop() 함수
exam_data = {'math' : [90, 80, 70], 'korea' : [98, 89, 95],
              'english' : [85,96,100], 'history' : [100, 90, 90]}
df = pd.DataFrame(exam_data, index = ['Kim', 'Lee', 'Park'])
df
```

```python
# drop('행 or 열 이름', axis = 0 or 1, inplace=True)

# Numpy :
#     axis = 0 : 열 / axis = 1 : 행
# Pandas :
#     axis = 0 : 행 / axis = 1 : 열

# 행을 삭제하려면 axis = 0 (기본값)
# 열을 삭제하려면 axis = 1 << 꼭 지정 해야함!~
df2 = df.copy()

df2.drop('Lee', inplace=True)
df2
```

```python
# 열 삭제
# 2개 이상의 열을 drop하고 싶으면 대괄호[] 안에 명시
df3 = df.copy()

df3.drop(['math', 'history'], axis = 1, inplace=True)
df3
```

# 데이터프레임 인덱싱 : loc/ iloc 함수

loc와 iloc는 가장 많이 사용되는 함수

데이터 전처리할 때 인덱싱이 가장 많이 사용되기 때문!

loc는 이름으로, iloc는 숫자로 인덱싱

```python
first = city.loc['A']
print(first)
second = city.iloc[1]
print(second)
```

```python
exam_data = {'math' : [90, 80, 70], 'korea' : [98, 89, 95],
              'english' : [85,96,100], 'history' : [100, 90, 90]}
df = pd.DataFrame(exam_data, index = ['Kim', 'Lee', 'Park'])
df
```

```python
# kim과 lee의 점수 출력

# 두 개 이상의 행을 선택할 때는 대괄호를 신경써야...!!
# 각각 선택은 두 개[[ ]] / 범위 ( : )로 선택은 한 개 []

df.loc[['Kim', 'Lee']]
df.loc['Kim':'Lee']
df.iloc[0:2]
df.iloc[[0,1]]
```

```python
# 열 선택
#   열 선택은 따로 함수가 필요 X
# df.열이름 or df['열이름']으로 선택이 가능
```

```python
# 수학 열만 선택
a= df.math
b=df['math']
print(a)
print()
print(b)
```

```python
# 범위 슬라이싱
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'],
             'math' : [90,80,70,60,50,40,30],
             'eng': [90,80,70,60,50,40,30],
             'kor' : [90,80,70,60,50,40,30],
             'pt' :[90,80,70,60,50,40,30]}
df = pd.DataFrame(exam_data)
df
```

```python
# 모든 행을 2행 간격으로 출력
df.iloc[0:7:2]
```

```python
# 0행부터 5행까지 2행 간격으로 출력
df.iloc[0:6:2]
```

```python
# 전체 데이터 역순 정렬 > 모든 행을 -1행 간격
df.iloc[7::-1]
```

```python
# 요소 선택
# set_index() 함수 : 특정한 열을 인덱스로 부여해주는 함수
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'],
             'math' : [90,80,70,60,50,40,30],
             'eng': [90,80,70,60,50,40,30],
             'kor' : [90,80,70,60,50,40,30],
             'pt' :[90,80,70,60,50,40,30]}
df = pd.DataFrame(exam_data)
df.set_index('name', inplace=True)
df
```

```python
# loc, iloc 행 / 열 선택과정에서
# 두 개 이상의 행이나 두 개 이상의 열을 선택하려면 [[ ]]을 이용
# ... 하나의 대괄호만 [] 쓴다면?..
df.loc['Ha','kor']
```

```python
# 해당하는 요소 하나를 가져오라는 의미 !!
df.iloc[3,2]
```

```python
# Ah의 eng, kor 점수를 위에서 한 4가지 방법을 이용해서 출력
df.loc['Ah',['eng','kor']]

df.loc['Ah','eng':'kor']

df.iloc[2,1:3]

df.iloc[2,[1,2]]
```

```python
# Oh, Woo의 kor, pt 점수를 4가지 방법으로 출력
df.loc[['Oh','Woo'],['kor','pt']]

df.iloc[[0,1],[2,3]]

df.loc['Oh':'Woo','kor':'pt']

df.iloc[0:2,2:4]
```

```python
# 행, 열 추가
# 열 추가는 df['새로운 열 이름'] = 값의 형태로 추가
# 새로운 열 이름이 아니라 기존 열 이름이라면..=> 그대로 덮어씌워짐!
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'],
             'math' : [90,80,70,60,50,40,30],
             'eng': [90,80,70,60,50,40,30],
             'kor' : [90,80,70,60,50,40,30],
             'pt' :[90,80,70,60,50,40,30]}
df = pd.DataFrame(exam_data)
df['music'] = 80
df
# 요소 값 변경하기
# df.loc[3] = 10 # 단일 값을 넣었다면 => 행 전체에는 모두 다 그 값으로 채워짐
# df.loc[4] = ['Wa', 100, 100, 100, 100, 100] # 행 또는 열의 수에 맞제 요소를 변경/추가
df.loc[6] = ['asdf', 50, 50, 50, 50, 50]
```

```python
# 행과 열 바꾸기 : 전치행렬 transpose() 함수 or .T
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'],
             'math' : [90,80,70,60,50,40,30],
             'eng': [90,80,70,60,50,40,30],
             'kor' : [90,80,70,60,50,40,30],
             'pt' :[90,80,70,60,50,40,30]}
df = pd.DataFrame(exam_data)
print(df)
print()

df = df.transpose()
print(df)
print()

df = df.T
print(df)
# 전치를 두 번하면 원상태로 돌아옴!!
```
